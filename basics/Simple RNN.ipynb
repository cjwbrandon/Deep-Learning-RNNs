# Create a simple RNN architecture to map

```python
class Simple_RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers, nonlinearity='tanh', bidirectional=False):
        super(Simple_RNN, self).__init__()
        
        # def rnn layer
        self.rnn = nn.RNN(input_size=input_size, hidden_size=hidden_size, num_layers=num_layers,
                          nonlinearity=nonlinearity, bidirectional=bidirectional)
        # def fc layer
        self.fc = nn.Linear(hidden_size, output_size)
        self.hidden_size = hidden_size # store to reshape later
        
    def forward(self, x, hidden):
        batch_size = x.size(0)
        
        # rnn layer
        x, hidden = self.rnn(x, hidden)
        x = x.view(-1, self.hidden_size) # flatten batch size and seq_len
        
        # fc layer
        output = self.fc(x)
        
        return output, hidden
```

```python
# hyperparameters
input_size=1 
output_size=1
hidden_dim=32
n_layers=1

# initialise the network
simple_rnn = Simple_RNN(input_size=input_size, hidden_size=hidden_dim, output_size=output_size,
                        num_layers=n_layers, nonlinearity='tanh', bidirectional=False)
print(simple_rnn)

# Loss & optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(simple_rnn.parameters(), lr=0.01)
```

```python
epochs = 100
seq_len = 100

train_loss = []
hidden = None
n_visualise = 20

for e in range(epochs):
    running_loss_train = 0 # training loss for this epoch
    
    # cosine wave
    x = np.linspace(e*np.pi, (5+e)*np.pi, seq_len+1)
    x = np.cos(x)
    x = x.reshape((seq_len+1, 1))
    
    y = x[1:]
    x = x[:-1]
    
    # convert to tensors
    x_tensor = torch.Tensor(x).unsqueeze(0) # add an additional dimension for batch size
    y_tensor = torch.Tensor(y)
    
    # zero grads
    simple_rnn.zero_grad()
    
    # feed into rnn
    preds, hidden = simple_rnn(x_tensor, hidden)
    hidden = hidden.data # update hidden
    
    # calculate loss
    loss = criterion(preds, y_tensor)
    running_loss_train += loss.item()*x_tensor.size(0)
    # perform backprop & update weights
    loss.backward()
    optimizer.step()
    
    train_loss.append(running_loss_train)
    
    print(f"Epochs: {int(e+1)}/{int(epochs)} --- Training loss: {running_loss_train}")
    
    # Visualise very n epochs
    if (e+1) % n_visualise == 0:
        plt.figure(figsize=(8,4))

        # visualise data
        plt.plot(range(seq_len), preds.data.numpy(), color='red', label='predictions')
        plt.plot(range(seq_len), y, color='blue', label='actuals')

        plt.legend(loc='best')
        plt.show()
```
